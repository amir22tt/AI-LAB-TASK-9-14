```python
import pandas as pd
df = pd.read_csv('practice_dataset.csv')
```

```python
print(df.isnull().sum())
```

**handling null values**

```python
avg_age = df['Age'].mean()
print(f"Average age: {avg_age:.2f}")
```

```python
df['Age'] = df['Age'].fillna(avg_age)
```

```python
avg_water = df['Water_Consumed_Liters'].mean()
print(f"Average water: {avg_water:.2f}")
```

```python
df['Water_Consumed_Liters'] = df['Water_Consumed_Liters'].fillna(avg_water)
```

```python
most_common_activity = df['Activity_Level'].mode()[0]
print(f"common Activity: {most_common_activity}")
```

```python
df['Activity_Level'] = df['Activity_Level'].fillna(most_common_activity)
```

```python
print(df.isnull().sum())
```

**changing datatype**

```python
print(df.dtypes)
```

```python
df['Age'] = df['Age'].astype('int')
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```
